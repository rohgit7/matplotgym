```python
import pandas as pd
from matplotlib import pyplot as plt

data=pd.read_csv('gym.csv')

data.head()
```

```python
newdata = data.sample(frac=0.1, random_state=1)
newdata.to_csv('sampled.csv', index=False)

datas= pd.read_csv('sampled.csv')
datas.head()
```

```python
plt.scatter(data['Age'],data['Calories_Burned'],color='red',alpha=0.25)
plt.title('Age group v/s Calories Burned')
plt.xlabel('Age')
plt.ylabel('Calories Burned')
plt.show()
```

```python
wtype = data.groupby(data['Workout_Type'])['Height (m)'].count()
wtype.plot(kind='bar', color = 'violet')
plt.title('Workout Type')
plt.ylabel('Count')
plt.show()
```

```python
maxb = data.groupby(data['Max_BPM'])['Age'].count()
avg= data.groupby(data['Avg_BPM'])['Age'].count()
rest= data.groupby(data['Resting_BPM'])['Age'].count()


plt.subplot(2,2,1)
plt.xlabel('Max_BPM')
plt.ylabel('Age')
maxb.plot(color= 'green',linewidth=1, ls='dashdot')

plt.subplot(2,2,2)
plt.xlabel('Average_BPM')
plt.ylabel('Age')
avg.plot( color= 'red', linewidth=1, ls=':')

plt.subplot(2,2,3)
plt.xlabel('Resting_BPM')
plt.ylabel('Age')
rest.plot(color= 'orange',linewidth=1)



plt.tight_layout()
plt.suptitle('Measure of BPM and weight at different ages', y='1.05')
plt.subplots_adjust(top=1)
plt.show()
```
